# Random walk

## What is Random Walk?

A time series said to follow a random walk if the first differences (difference from one observation to the next observation) are random.

**The current observation in a random walk is a random step from previous observation**. This property makes a random walk time dependent and hence non-stationary.

A random walk is unpredictable; it cannot reasonably be predicted.

```python
#Importing libraries

import warnings
import itertools
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
plt.style.use('fivethirtyeight')

import pandas as pd
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.ar_model import AR
from statsmodels.tsa.arima_model import ARMA, ARIMA
from pmdarima.arima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
#from fbprophet import Prophet

from math import sqrt

import matplotlib
matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['text.color'] = 'k'
import seaborn as sns

from random import random

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, median_absolute_error, mean_squared_log_error
```

Forecast evaluation function (from previous notebook)

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluate_forecast(y,pred):
    results = pd.DataFrame({'r2_score':r2_score(y, pred),
                           }, index=[0])
    results['mean_absolute_error'] = mean_absolute_error(y, pred)
    results['median_absolute_error'] = median_absolute_error(y, pred)
    results['mse'] = mean_squared_error(y, pred)
    #results['msle'] = mean_squared_log_error(y, pred)
    results['mape'] = mean_absolute_percentage_error(y, pred)
    results['rmse'] = np.sqrt(results['mse'])
    return results
```

 ## Random walk data generation and prediction

```python
## generate random walk data
from random import random
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
import pandas as pd

random_walk = list()
index = pd.date_range('1/1/2000', periods=1000, freq='D')
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)

random_walk = pd.Series(random_walk, index=index)
random_walk.plot(figsize=(14, 7))
plt.show()
```

Given the way that the random walk is constructed, we can expect that the best prediction we could make would be to use the observation at the previous time. This is because the next step will be a function of the previous step

This is often called the naive forecast, or a persistence model.

```python
# prepare dataset
train_size = int(len(random_walk) * 0.66)
train, test = random_walk[0:train_size], random_walk[train_size:]

# persistence model
predictions = list()
history = train[-1]
for i in range(len(test)):
    yhat = history
    predictions.append(yhat)
    history = test[i]

evaluate_forecast(test, predictions)    
```

```python
## visualize forecasted & observed values
predictions = pd.Series(predictions, index=test.index)
ax = random_walk.plot(label='observed')
predictions.plot(ax=ax, label='Forecast', alpha=.7, figsize=(14, 7))
ax.set_xlabel('Date')
plt.legend()
plt.show()
```

# Unobserved component modeling

## What is an unobserved components model?

An unobserved components model decomposes time series into components such as trend, season, cycle, regression effects due to predictors and irregular component

Mathematically a UCM model looks as follows:

  yt  =t + t + t + jxjt + t 

where the components t , t, and t represent the trend, seasonal, and cyclical components, respectively; the term mj=1jxjt gives the contribution of regression variables with fixed or time varying regression coefficients.


## When do we use an unobserved components model?

A UCM can be used when we want to decompose the time series into time series components such as trend, seasonality, cyclicity and also include other exogenous variables




```python
from IPython.display import display, Latex
from pandas_datareader.data import DataReader

# Get the raw data
start = '1948-01'
end = '2008-01'
us_gnp = DataReader('GNPC96', 'fred', start=start, end=end)
us_gnp_deflator = DataReader('GNPDEF', 'fred', start=start, end=end)
us_monetary_base = DataReader('AMBSL', 'fred', start=start, end=end).resample('QS').mean()
recessions = DataReader('USRECQ', 'fred', start=start, end=end).resample('QS').last().values[:,0]

# Construct the dataframe
us_data = pd.concat(map(np.log, (us_gnp, us_gnp_deflator, us_monetary_base)), axis=1)
us_data.columns = ['US GNP','US Prices','US monetary base']
us_dates = us_data.index._mpl_repr()
```

Here, the data is already seasonally adjusted and there are no obvious explanatory variables

* The irregular will be assumed to be white noise, and the cycle will be stochastic and damped. The final modeling choice is the specification to use for the trend component. Harvey and Jaeger consider two models:

    * Local linear trend (the "unrestricted" model)
    * Smooth trend (the "restricted" model, since we are forcing  =0 )
    
    
Below, we construct kwargs dictionaries for each of these model types. Notice that rather that there are two ways to specify the models. One way is to specify components directly, as in the table above. The other way is to use string names which map to various specifications.

```python
#creating the train and validation set
from sklearn.model_selection import TimeSeriesSplit

train = us_data[:int(0.8*(len(us_data)))]
valid = us_data[int(0.8*(len(us_data))):]
```

```python
## model data
model = sm.tsa.UnobservedComponents(train['US GNP'], 'llevel', cycle=True, stochastic_cycle=True)
results = model.fit()
fig = results.plot_components(figsize = (30,20))
```

```python
# Check for performance
expected = valid['US GNP']
predictions = results.forecast(steps=len(valid))
forecast_errors = [expected[i]-predictions[i] for i in range(len(expected))]
#print('Forecast Errors: %s' % forecast_errors)
bias = sum(forecast_errors) * 1.0/len(expected)
print('Bias: %f' % bias)
```

```python
evaluate_forecast(expected, predictions)  
```

# Multivariate Time Series

## What  is multivariate time series?
A Multivariate time series has more than one time-dependent variable. Each variable depends not only on its past values but also has some dependency on other variables. This dependency is used for forecasting future values

## When do we use multivariate time series?

Multivariate time series forecasting is used when data has multiple time dependent factors

Let's try to understand this with the help of an example

### Introduction to VAR
One of the most commonly used methods for multivariate time series forecasting is Vector Auto Regression (VAR).
In a VAR model, each variable is a linear function of the past values of itself and the past values of all the other variables.

For example we have two variables, y1 and y2. We need to forecast the value of these two variables at time t, from the given data for past n values.

```python
## load data 
df = pd.read_csv("./Data/AirQualityUCI.csv", parse_dates=[['Date', 'Time']])

df = df.drop(['Unnamed: 15','Unnamed: 16'], axis=1)

df.head()
```

The data type of the Date_Time column is object and we need to change it to datetime. Also, for preparing the data, we need the index to have datetime. Follow the below commands:

```python
df['Date_Time'] = pd.to_datetime(df.Date_Time , format = '%m/%d/%Y %H:%M:%S')
```

```python
df.head()
```

```python
## create Date_Time index
data = df.drop(['Date_Time'], axis=1)
data.index = df.Date_Time
data.head()
```

__Missing value treatment__
The next step is to deal with the missing values. 
You can choose to substitute the value using the average of a few previous values, or the value at the same time on the previous day

```python
#missing value treatment Since the missing values in the data are replaced with a value -200,
#we will have to impute the missing value with a better number.
cols = data.columns
for j in cols:
    for i in range(0,len(data)):
        if data[j][i] == -200:
            data[j][i] = data[j][i-1]
            
```

#### Stationarity of multivariate time series
Similar to the Augmented Dickey-Fuller test for univariate series, we have Johansens test for checking the stationarity of any multivariate time series data.

```python
#checking stationarity
from statsmodels.tsa.vector_ar.vecm import coint_johansen

#since the test works for only 12 variables, we have randomly dropped
#in the next iteration, we would drop another and check the eigenvalues

johan_test_temp = data.drop(['CO(GT)'], axis=1)
johan_test_temp.shape
```

```python
## showing eigen values for each feature
coint_johansen(johan_test_temp,-1,1).eig
```

```python
## Critical values (90%, 95%, 99%) for maximum eigenvalue statistic
coint_johansen(johan_test_temp,-1,1).cvm
```

Since the eigen value for each feature do not exceed the critical values anywhere, the null hypothesis that time series are not cointegrated cannot be rejected.

We can now go ahead and create the validation set to fit the model, and test the performance of the model:



#### Train Validation Split
The idea of creating a validation set is to analyze the performance of the model before using it for making predictions.
__How to do it__:
1. One cannot directly use the train_test_split or k-fold validation since this will disrupt the pattern in the series. The validation set should be created considering the date and time values
2. Suppose we have to forecast the temperate, dew point, cloud percent, etc. for the next two months using data from the last two years. One possible method is to keep the data for the last two months aside and train the model on the remaining 22 months

```python
#creating the train and validation set
train = data[:int(0.8*(len(data)))]
valid = data[int(0.8*(len(data))):]
```

#### Train & Fit the model

```python
#train the model
from statsmodels.tsa.vector_ar.var_model import VAR

model = VAR(endog=train)
```

```python
# fit the model & make predictions
model_fit = model.fit()
prediction = model_fit.forecast(model_fit.y, steps=len(valid))
```

#### Check performance on validation data

```python
#converting predictions to dataframe
pred = pd.DataFrame(index=range(0,len(prediction)),columns=[cols])
for j in range(0,13):
    for i in range(0, len(prediction)):
        pred.iloc[i][j] = prediction[i][j]
```

```python
pred.head()
```

```python
#check rmse
for i in cols:
    print('RMSE value for', i, 'is : ', sqrt(mean_squared_error(pred[i], valid[i])))
```

The other evaluation metrics like MAPE, MAE, MSE etc., can also be calculated in a similar way.
